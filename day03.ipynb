```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.master("local[*]") \
    .enableHiveSupport().appName("ex01_1") \
    .getOrCreate()

spark.version   # 2.4.5
```

```python
spark = SparkSession.builder.master("local[*]").appName("myApp") \
    .enableHiveSupport().appName("hive01") \
    .config("spark.mongodb.input.uri", "mongodb://127.0.0.1:27017/db1.t01") \
    .config("spark.mongodb.output.uri", "mongodb://127.0.0.1:27017/db1.t02") \
    .config('spark.jars.packages', 'org.mongodb.spark:mongo-spark-connector_2.11:2.4.1') \
    .getOrCreate()

# mongodb에서 읽기
df = spark.read.format("com.mongodb.spark.sql.DefaultSource").load()
df.show()
```

```python
df.createOrReplaceTempView("table1")
spark.sql("SELECT * FROM table1").show()
```

```python
df1=df.toPandas()
df1
```

# 1. 평균온도를 tmn_tmx로 추가

```python
df1 = df1.astype({'tmn': 'int'})
df1 = df1.astype({'tmx': 'int'})

df1['tmn_tmx']=(df1['tmn']+df1['tmx'])/2
df1
```

```python
spark.sql("select *, (tmn+tmx)/2 as 'tmn_tmx' from table1").show()
spark.sql("select * from table1").show()
```

# 4. 목요일 가장 온도가 낮은 지역은?

```python
df1['tmEf'] = pd.to_datetime(df1['tmEf'],format='%Y-%m-%d %H:%M:%S', errors='raise')
df1['date'] = df1['tmEf'].dt.day_name()
df1
```

```python
tmin=df1[df1['date']=="Thursday"]['tmn'].min()
df1[df1['tmn']==tmin]
df1[df1['date']=='Thursday']
date1=(df1['date']=="Thursday") & (df1['tmn']==tmin)
df1[date1]
```

# 5. 금요일 가장 온도가 높은 지역은?

```python
tmax=df1[df1['date']=="Friday"]['tmx'].max()
df1[df1['tmx']==tmax]
df1[df1['date']=='Friday']
date2=(df1['date']=="Friday") & (df1['tmx']==tmax)
df1[date2]
```
