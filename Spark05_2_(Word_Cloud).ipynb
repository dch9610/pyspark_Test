```python
# 자료 받기
import wikipedia
page = wikipedia.page("Samsung") # 위키피디아에 찾고 싶은 단어를 검색
text = page.content
text
```

```python
# 공백으로 구분 한 후 단어별 카운트
from collections import Counter
text1 = text.split(" ") # 공백으로 나눔
tag_counts = Counter(text1)
```

```python
# 상위 10개 단어 표시
tag_counts.most_common(10)
```

```python
# 폰트 설정
import matplotlib.pyplot as plt
import platform

print(platform.system())
if platform.system() == 'Windows':   #윈도우의 경우
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin":   #Mac 의 경우
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
elif platform.system() == "Linux":  # linux인 경우
    font_path ="/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
```

```python
from wordcloud import WordCloud
wordcloud=WordCloud(font_path= font_path, 
                    background_color="white",
                    max_words=100,
                    relative_scaling= 0.5,  #최소 0부터 최대 1까지 값 지정.
                    width = 800,
                    height = 400
                 ).generate_from_frequencies(tag_counts)  
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('tag-wordcloud.png') 
```

```python
# 전치사 제거
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(font_path= font_path, 
                    stopwords = set(STOPWORDS),
                    background_color="white",
                    max_words=100,
                    relative_scaling= 0.5,  #최소 0부터 최대 1까지 값 지정.
                    width = 800,
                    height = 400
                 ).generate(text)  
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('tag-wordcloud.png') 
```
