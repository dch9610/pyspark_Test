```python
import pandas as pd

jeju = pd.read_excel('./files4/3_1_crawling_raw.xlsx')
jeju = jeju[['place','tags']]
jeju
```

```python
# 단어별 카운트
hashtag = []
for i in jeju['tags']:
    text1 = i.split("'")
    for j in text1:
        if (len(j) >=3):
            hashtag.append(j[1:])
hash1 = []
for i in hashtag:
    if ((i.find('제주') != 0) & (i.find('서귀포')!=0) ) :
        hash1.append(i)
print(type(hash1))
hash1
```

```python
from collections import Counter
tag_counts = Counter(hash1)

df1 = pd.DataFrame(list(tag_counts.most_common(100)), columns=['name_official','count'])
df1.head(30)
```
